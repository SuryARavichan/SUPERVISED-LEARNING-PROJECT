```python
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
yes_churn
```

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
```

```python
no_churn
```

```python
churn_rate = 100*yes_churn / (no_churn + yes_churn)
```

```python
churn_rate.transpose()['customerID'].plot()
```

```python
churn_rate.transpose()['customerID']
```

```python
def cohort(tenure):
  if tenure < 13:
    return '0-12 Months'
  elif tenure < 25:
    return '12-24 Months'
  elif tenure < 49:
    return '24-48 Months'
  else:
    return 'Over 48 Months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort','tenure']]
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort',alpha=0.4,linewidth=0.3,palette='RdGy_r')
```

```python
sns.catplot(data=df,x='Tenure Cohort',hue='Churn',kind='count',col='Contract')
```
